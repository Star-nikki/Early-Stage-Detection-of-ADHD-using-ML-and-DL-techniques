```python
import nibabel as nib
import os
import numpy as np

files = os.listdir(data_path)

# Define functions for getting various statistics
def get_mean(file_path):
    nii_file = nib.load(file_path)
    return nii_file.get_fdata().mean()

def get_median(file_path):
    nii_file = nib.load(file_path)
    return np.median(nii_file.get_fdata())

def get_std(file_path):
    nii_file = nib.load(file_path)
    return np.std(nii_file.get_fdata())

def get_var(file_path):
    nii_file = nib.load(file_path)
    return np.var(nii_file.get_fdata())

# Analyze one file at a time
max_mean = -np.inf
max_median = -np.inf
max_std = -np.inf
max_var = -np.inf
max_mean_file = ''
max_median_file = ''
max_std_file = ''
max_var_file = ''

for file_name in files:
    file_path = os.path.join(data_path, file_name)
    
    # Compute various statistics for the current file
    mean_value = get_mean(file_path)
    median_value = get_median(file_path)
    std_value = get_std(file_path)
    var_value = get_var(file_path)
    
    # Update the maximum values and filenames
    if mean_value > max_mean:
        max_mean = mean_value
        max_mean_file = file_name
    if median_value > max_median:
        max_median = median_value
        max_median_file = file_name
    if std_value > max_std:
        max_std = std_value
        max_std_file = file_name
    if var_value > max_var:
        max_var = var_value
        max_var_file = file_name
    
    # Print the statistics for the current file
    print(f'File: {file_name}')
    print(f'Mean: {mean_value:.2f}')
    print(f'Median: {median_value:.2f}')
    print(f'Std: {std_value:.2f}')
    print(f'Var: {var_value:.2f}')
    print('----------------------')
    
# Print the filenames with the maximum values
print(f'File with highest mean: {max_mean_file} ({max_mean:.2f})')
print(f'File with highest median: {max_median_file} ({max_median:.2f})')
print(f'File with highest std: {max_std_file} ({max_std:.2f})')
print(f'File with highest var: {max_var_file} ({max_var:.2f})')
```

### Temporal normalization to each 4D NIfTI file, normalizing the temporal data within the range [0, 1]

```python
# Define the temporal normalization function
def temporal_normalization(file_path):
    nii_file = nib.load(file_path)
    data = nii_file.get_fdata()

    # Determine the index of the temporal dimension
    temporal_dim_index = data.shape.index(data.shape[-1])

    # Extract the temporal data
    temporal_data = np.take(data, indices=3, axis=temporal_dim_index)

    # Perform temporal normalization
    min_value = np.min(temporal_data)
    max_value = np.max(temporal_data)

    # Linear scaling to the range [0, 1]
    normalized_data = (temporal_data - min_value) / (max_value - min_value)

    # Replace the temporal slice in the original 4D volume with the normalized data
    data = np.insert(data, 3, normalized_data, axis=temporal_dim_index)

    # Create a new NIfTI image with the normalized data
    normalized_nii = nib.Nifti1Image(data, nii_file.affine, nii_file.header)

    return normalized_nii

# Iterate over the files and perform temporal normalization on each one
for file_name in files:
    file_path = os.path.join(data_path, file_name)
    normalized_nii = temporal_normalization(file_path)

    # Save the normalized NIfTI image to a new file
    normalized_file_name = 'normalized_' + file_name
    normalized_file_path = os.path.join(data_path, normalized_file_name)
    nib.save(normalized_nii, normalized_file_path)
```
